```python
import matplotlib.pyplot as plt
from PIL import Image

# Open the JPG image
image_path = '/content/pic1.jpg'  # Replace with the path to your JPG file
img = Image.open(image_path)

# Convert the image to grayscale (optional)
img_gray = img.convert('L')

# Calculate the histogram
histogram = img_gray.histogram()

# Split the histogram into its color channels
red_histogram = histogram[0:256]
green_histogram = histogram[256:512]
blue_histogram = histogram[512:768]

# Plot the histograms
plt.figure(figsize=(12, 6))
plt.subplot(131)
plt.title('Red Histogram')
plt.bar(range(256), red_histogram, color='red', alpha=0.7)
plt.subplot(132)
plt.title('Green Histogram')
plt.bar(range(256), green_histogram, color='green', alpha=0.7)
plt.subplot(133)
plt.title('Blue Histogram')
plt.bar(range(256), blue_histogram, color='blue', alpha=0.7)

plt.tight_layout()
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt

# Load the JPG image
image_path = '/content/pic1.jpg'  # Replace with the path to your JPG image
image = cv2.imread(image_path)

# Convert the image to grayscale (optional)
gray_image = cv2.cvtColor(image, cv2.COLOR_BGR2GRAY)

# Calculate the histogram
histogram = cv2.calcHist([gray_image], [0], None, [256], [0, 256])

# Plot the histogram
plt.figure(figsize=(8, 6))
plt.title('Histogram of the Image')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.plot(histogram)
plt.xlim([0, 256])  # Set the x-axis range to [0, 256] for grayscale images
plt.show()
```

```python
import cv2
import matplotlib.pyplot as plt
import numpy as np

# Load the first JPG image
image_path1 = '/content/pic2 - Copy.jpg'  # Replace with the path to your first JPG image
image1 = cv2.imread(image_path1)

# Load the second JPG image
image_path2 = '/content/pic2.jpg'  # Replace with the path to your second JPG image
image2 = cv2.imread(image_path2)

# Convert both images to grayscale (optional)
gray_image1 = cv2.cvtColor(image1, cv2.COLOR_BGR2GRAY)
gray_image2 = cv2.cvtColor(image2, cv2.COLOR_BGR2GRAY)

# Calculate histograms for both images
histogram1 = cv2.calcHist([gray_image1], [0], None, [256], [0, 256])
histogram2 = cv2.calcHist([gray_image2], [0], None, [256], [0, 256])

# Calculate a similarity score (e.g., Chi-Squared distance) between histograms
# Lower scores indicate higher similarity
score = cv2.compareHist(histogram1, histogram2, cv2.HISTCMP_CHISQR)

# Define a threshold for similarity (you can adjust this threshold)
threshold = 0.2

# Compare the histograms and print the result
are_images_same = score < threshold
print("Are the images the same?", are_images_same)

# Plot histograms side by side
plt.figure(figsize=(12, 6))

# Plot histogram of Image 1
plt.subplot(121)
plt.title('Histogram of Image 1')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.plot(histogram1, color='blue')
plt.xlim([0, 256])  # Set the x-axis range to [0, 256] for grayscale images

# Plot histogram of Image 2
plt.subplot(122)
plt.title('Histogram of Image 2')
plt.xlabel('Pixel Value')
plt.ylabel('Frequency')
plt.plot(histogram2, color='red')
plt.xlim([0, 256])  # Set the x-axis range to [0, 256] for grayscale images

plt.tight_layout()
plt.show()
```
