```python
import matplotlib.pyplot as plt
import numpy as np
x=np.linspace(0,1,10)
m=-0.6
c=25
y=m*x+c
plt.plot(x,y)
plt.show()
```

```python
import matplotlib.pyplot as plt
import numpy as np
plt.style.use('_mpl-gallery')

x=np.linspace(0,10,100)
y=np.sin(x)
fig,a=plt.subplots()
a.plot(x,y,linewidth=3)
a.set(xlim=(0.8),xticks=np.arrange(1,9),
      ylim=(0.8),yticks=np.arrange(1,9))
plt.show()
```

```python
#temperature dependance of resistance for thermistor
import matplotlib.pyplot as plt
import numpy as np

Temp=np.array([10,20,25,30,40,50,60,70,80,90,100,110,120])
Res=np.array([18476,12185,10000,8260,5740,4080,2960,2188,1645,1257,973,765,608])

fig,a=plt.subplots()
a.plot(Temp,Res,linewidth=2)
plt.scatter(Temp,Res)

plt.show()

plt.scatter(Temp,Res)

fig,a=plt.subplots()
a.plot(Temp,Res,linewidth=3)

```
