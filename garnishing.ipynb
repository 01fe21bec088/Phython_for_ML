```python
#temperature dependance of resistance for thermistor
import matplotlib.pyplot as plt
import numpy as np

Temp=np.array([10,20,25,30,40,50,60,70,80,90,100,110,120])
Res=np.array([18476,12185,10000,8260,5740,4080,2960,2188,1645,1257,973,765,608])
plt.plot(Temp,Res)
plt.show()

plt.plot(Res, marker='*')
plt.show()

plt.plot(Res, linestyle='dotted')
plt.show()

plt.plot(Temp,Res)
plt.title("thermistor")
plt.xlabel("Temp")
plt.ylabel("Res")
plt.show()

plt.plot(Temp,Res)
plt.grid()
plt.show()
```
